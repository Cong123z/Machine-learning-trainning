## Dữ liệu 

Tập dữ liệu giá nhà ở Boston đã có sẵn trên sklearn, dữ liệu đã được chuẩn hóa và chia thành tập train, tập test

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import math 

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

# Đọc dữ liệu 

Dữ liệu về giá nhà ở Boston được hỗ trợ bởi sklearn, đọc dữ liệu thông qua hàm `datasets.load_boston()` 

Xem thêm các bộ dữ liệu khác tại https://scikit-learn.org/stable/datasets/index.html#toy-datasets. 

Dữ liệu được chia thành các thành phần data và target như tập diabetes. Dữ liệu cũng đã được chuẩn hóa, chỉ cần gọi ra và huấn luyện

```python
# lay du lieu dataset - du lieu ve giá nhà
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("BostonHousing.csv")    
df.head()
```

**Chia dữ liệu làm 2 phần training 404 mẫu và testing 102 mẫu** 

```python
# cat nho du lieu, lay 1 phan cho qua trinh thu nghiem,
# chia train test cac mau du lieu
X=df.drop(['medv'],axis=1)
Y=df['medv']
X=(X-X.mean())/(X.max()-X.min())
X.shape
train_size=0.8
x_train = X[:(int)(len(X)*train_size)]
y_train = Y[:(int)(len(X)*train_size)]
x_test = X[(int)(len(X)*train_size):]
y_test = Y[(int)(len(X)*train_size):]
x_train.info()
print(f"Số lượng train và test là : {len(x_train)} và {len(x_test)}")

```

# 3.Xây dựng mô hình

## 3.1.Xây dựng mô hình bằng thư viện


```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

## 3.2.Xây dựng mô hình Linear Regression tự viết 

```python

one_col=np.ones((len(x_train),1))
A=np.concatenate((one_col,x_train),axis=1)#Tạo ma trận A m*(n+1)
print(A[:5])

A_T=np.transpose(A)#Ma trận chuyển vị A
w=np.dot(A_T,A)
w=np.linalg.inv(w)#Ma trận nghịch đảo
w=np.dot(w,np.dot(A_T,y_train))
w.shape
```

## 3.3.Hàm test mô hình tự viết

```python
def predict_model(w,x_test):
    bias=w[0]
    coef=w[1:]
    coef=coef.reshape(len(coef),1)
    y_pred=np.dot(x_test,coef)+bias
    return y_pred
```

# 4.Huấn luyện mô hình

## 4.1.Huấn luyện mô hình của thư viện

```python
model=reg.fit(x_train,y_train)
print(model.get_params())
```

## 4.2.Training mô hình bằng Linear regression tự viết

```python
y_pred=predict_model(w,x_test)
y_pred=y_pred.reshape(-1)
y_pred.shape
```

# 5.Dự đoán các mẫu dữ liệu

## 5.1.Dự đoán các mẫu dữ liệu theo mô hình của thư viện

```python
y_pred1=model.predict(x_test)
pd.DataFrame(data=np.array([y_test, y_pred1,abs(y_test - y_pred1),abs(y_test-y_pred1)*100/y_test]).T,columns=["Thực tế", "Dự đoán", "Lệch","Lệch theo %"])
```

## 5.2.Dự đoán các mẫu dữ liệu tính theo linear regression tự viết

```python
pd.DataFrame(data=np.array([y_test, y_pred,abs(y_test - y_pred),abs(y_test-y_pred)*100/y_test]).T,columns=["Thực tế", "Dự đoán", "Lệch","Lệch theo %"])
```

## 5.3.Đánh giá mô hình linear regression của thư viện

```python
print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred1)))
print('R2 Score:', r2_score(y_test, y_pred1))
```

## 5.4.Đánh giá mô hình linear regression tự viết

```python

print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R2 Score:', r2_score(y_test, y_pred))
```
